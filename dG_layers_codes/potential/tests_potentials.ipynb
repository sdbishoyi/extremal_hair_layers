# Test behavior of Potentials

```matlab
x_array = linspace(-10,10,1000);
test_omega = 1 - ((x_array + 5)./(-10 + 5)).^4;
y_array = x_array./test_omega;
plot(x_array,y_array)
```

```matlab
y_array(1)
y_array(2)
```

```matlab
test_omega = 1 - ((x_array + 5)./(-10 + 5)).^4.*heaviside(-5 - x_array) - ((x_array - 5)./(10 - 5)).^4.*heaviside(x_array - 5);
y_array = x_array./test_omega;
plot(x_array,y_array)
```
